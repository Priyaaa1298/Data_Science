```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings("ignore")  # Suppress all warnings
```

```python
df = pd.read_csv(r'C:\Users\Admin\Desktop\data science assighnments\Random Forest\Random Forest\random forest\data.csv')
```

```python

# Drop the 'Id' column as it's not needed for classification
df.drop("Id number", axis=1, inplace=True)
```

```python

# Split the features and target variable
X = df.iloc[:, :-1]  # Features
y = df["Type of glass"]       # Target variable
```

```python

# Standardize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = rf_classifier.predict(X_test)
```

```python

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

```python
# Apply BaggingClassifier
bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(), 
    n_estimators=50, 
    random_state=42
)
bagging.fit(X_train, y_train)

```

```python

# Predict and evaluate
y_pred = bagging.predict(X_test)
print("Bagging Classifier Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Apply AdaBoostClassifier
adaboost = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1), 
    n_estimators=50, 
    learning_rate=1.0, 
    random_state=42
)
adaboost.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred_boost = adaboost.predict(X_test)
print("AdaBoost Classifier Accuracy:", accuracy_score(y_test, y_pred_boost))
print("\nClassification Report:\n", classification_report(y_test, y_pred_boost))
```

Question Answers:-

1. Bagging creates multiple models (learners) by training each model independently on a randomly sampled subset of the training data, with replacement (bootstrap sampling).
After training, the predictions of these models are combined (typically by averaging for regression or majority voting for classification) to make a final prediction.

Boosting trains multiple models sequentially, with each new model focusing on correcting the mistakes made by the previous model.
Each model is trained with a re-weighted dataset, where incorrectly predicted samples from the previous model are given higher weights.

2. we can overcome imbalance data with help of resampling techniques. like  Oversampling the Minority Class, Undersampling the Majority Class, Combination of Over-sampling and Under-sampling (Hybrid). or with the help of ensemble methods or evaluation metrices.
