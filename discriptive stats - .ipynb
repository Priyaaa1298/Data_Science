```python
df_cleaned = df.drop(columns=['Date', 'Day','SKU','City','Volume','BU'])
```

```python
df_cleaned
```

```python
month_counts = df_cleaned['Brand'].value_counts()
plt.figure(figsize=(10, 6))
month_counts.plot(kind='barh', color='darkred')
plt.xlabel('Frequency')

plt.show()
```

```python
# Select numeric columns only
numeric_df = df_cleaned.select_dtypes(include='number')
```

```python
# Apply zscore to all numerical columns


z_scores = (numeric_df - numeric_df.mean()) / numeric_df.std()
print("Original DataFrame:")
print(numeric_df)
print("\nZ-score DataFrame :")
print(z_scores)
```

```python
# One-hot encoding use to encode all columns data in 0's and 1's format
df_encoded = pd.get_dummies(df_cleaned, columns=['Brand','Model'])
print(df_encoded)
```

```python
# Proportion of data within [-1, 1] for each column
proportion_within_range = ((z_scores >= -3) & (z_scores <= 3)).mean()
print(proportion_within_range)
```
