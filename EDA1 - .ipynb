```python
# Replace missing values with the median of each column
df.fillna(df.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 9))  # Set the figure size
sns.boxplot(data=df)  # Create the boxplot for all columns in the DataFrame

```

```python
# Create a pairplot
selected_columns = ['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP']
# Create a pairplot for the selected columns
sns.pairplot(df[selected_columns])
plt.show(5)
sns.pairplot(df[selected_columns], kind='kde')  # KDE plots on the diagonal
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 9))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Title for the plot
plt.title('Correlation Heatmap for Multiple Columns', fontsize=15)
plt.show()
```
